```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### 01 Plotting

```python
# Plotting with line (default)
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y)
plt.show()
```

```python
# Plotting without line
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x,y, 'o')
plt.show()
```

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x,y)
plt.show()
```

```python
# Default x points
y = np.array([3, 8 , 1, 10])

plt.plot(y)
plt.show()
```

### 02 Markers

```python
y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'o') # specify the marker parameter
plt.show()
```

```python
# marker | line | color
# format string fmt
y = np.array([3, 8, 1, 10])

plt.plot(y, 'o:r')
plt.show()
```

```python
# Marker Size 
y = np.array([3,8,1,10])

plt.plot(y, marker = 'o', ms = 20) # marker size (ms) parameter
plt.show()
```

```python
# Marker Edge Color
y = np.array([3,8,1,10])

plt.plot(y, marker = 'o', ms = 20, mec = 'r') # marker edge color (mec) parameter
plt.show()
```

```python
# Marker Face Color
y = np.array([3,8,1,10])

plt.plot(y, marker = 'o', ms = 20, mfc = 'r') # marker face color (mfc) parameter
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker = 'o', ms = 20, mec = 'r', mfc = 'y') # parameter
plt.show()
```

### 03 Line

```python
# line style
# dotted : | dashed -- | solid - | dashdot -. | none ''or ' '
y = np.array([3, 8, 1, 10])
plt.plot(y, ls = 'dotted') # linestyle (ls) parameter
plt.show()
```

```python
# Line color
y = np.array([3, 8, 1, 10])
plt.plot(y, c = 'r') # color (c) parameter
plt.show()
```

```python
# Line width
y = np.array([3, 8, 1, 10])
plt.plot(y, lw = 10) # line width (lw) parameter
plt.show()
```

```python
# Multiple lines
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
# Multiple line by order
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

### 04 Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y, c = 'r')
plt.title("Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y, c = 'r')
plt.title("Title", fontdict = font1, loc = 'left') # loc parameter: left | right | center
plt.xlabel("X axis",  fontdict = font2)
plt.ylabel("Y axis", fontdict = font2)
plt.show()
```

```python
# Add Title and SuperTitle
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES") # Add subplot title

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME") # Add subplot title

plt.suptitle("MY SHOP") # Add Super Title
plt.show()
```

### 07 Scatter Plots

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'r')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = 'b')

plt.show()
```

```python
# color map, size, alpha
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis', s = sizes, alpha = 0.6)
plt.colorbar()
plt.show()
```

### 08 Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y) # display bar horizontally
plt.show()
```

```python
# vertical bar uses width 
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
# horizontal bar uses height
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

### 09 Histogram

```python
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

### 10 Pie charts

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True) # start angle default at 0 and display counterclockwise
plt.legend(title = "Fruits") # Add legend
plt.show() 
```
